## Предобработка данных

### Обработка пропусков

Пойдём по порядку и будем рассматривать столбцы с пропущенными значениями. В зависимости от влияния того или иного параметра на цель исследования, будем принимать решение, на что заменять пропущенные значения или не заменять их вовсе.

#### Данные balcony

```python
# Выведем количество уникальных значений столбца
df['balcony'].value_counts()                 
```

```python
# Подсчитаем количество пропусков в столбце

print('В столбце balcony пропущено значений:', df['balcony'].isna().sum())
```

Вероятнее всего, столбец balcony заполнялся вручную, некоторые пользователи не указали количество балконов имея ввиду то, что балконов в помещении нет. Заменим пропуски в данном столбце на нули.

```python
# Подсчитаем количество пропусков в столбце
print('В столбце kitchen_area пропущено значений:', df['kitchen_area'].isna().sum())
```

Для замены пропущенных значений столбца сделаем следующее: найдем медианные значения total_area и kitchen_area c группировкой по количеству комнат. Затем, найдем соотношение площади кухни к общей площади квартиры. После этого найденные коэфициенты добавим к основной таблице, переумножим на значение общей площади и заменим получившимся значением пропуски в столбце kitchen_area.

```python
# находим медианные значения по total_area и kitchen_area
total = df.pivot_table(index='rooms', values='total_area', aggfunc='median')
kitchen = df.pivot_table(index='rooms', values='kitchen_area', aggfunc='median')
```

```python
# объединим медианы в одну таблицу
total_new = total.merge(kitchen, on='rooms')
```

```python
# найдем коэффиент отношения медианы площади кухни к медиане площади всей квартиры с по комнатам
total_new['share'] = total_new['kitchen_area'] / total_new['total_area']

# создадим датафрейм с долями площади кухни
total_new_1 = pd.DataFrame({'total_new_1': total_new['share']}).reset_index()

# присоединим полученные коффициенты к основной таблице
df = df.merge(total_new_1, on='rooms')

# Заменим пропуски в столце kitchen_area на произведение коэф-та с общей площадью квартиры
df['kitchen_area'] = df['kitchen_area'].fillna(value = (df['total_area'] * df['total_new_1']))
```

```python
# Подсчитаем количество пропусков в столбце
print('В столбце kitchen_area пропущено значений:', df['kitchen_area'].isna().sum())
```

### День недели, месяц и год публикации объявления

Для подсчёта зависимости между стоимости квартиры и датой размещения, переведем формат даты в datetime64 и выведем первые 5 строк датафрейма.

```python
df['first_day_exposition'] = pd.to_datetime(
    df['first_day_exposition'], format='%Y.'
)
display(df.head()) 
```

Теперь, когда у нас есть полноценная дата размещения, выведем из нее в отдельные столбцы день недели размещения, месяц и год. 

```python
# для этого напишем функцию, которая выделяет год из даты
def year(dat):
    for i in [dat]:
        return i.year

# применим функцию в новом столбце year 
df['year'] = df['first_day_exposition'].apply(year)

# функцию, которая выделяет месяц из даты
def month(dat):
    for i in [dat]:
        return i.month

# применим функцию в новом столбце month  
df['month'] = df['first_day_exposition'].apply(month)

# функцию, которая выделяет день недели из даты
def weekda(dat):
    for i in [dat]:
        return i.weekday()
    
# применим функцию в новом столбце day  
df['day'] = df['first_day_exposition'].apply(weekda)
```

```python
# выведем первые строки датафрейма
df.head()
```

В результате мы получили три столбца: год, месяц и день недели объявления.

### Расчёт этажа квартиры.

Теперь определим, на каком этаже находится квартира: первый, последний или другой. Для этого напишем функцию, которая будет считывать общее количество этажей в доме и этаж, на котором находится квартира.

```python
# напишем функцию, которая будет возвращать этаж в зависимости от значения floors_total & floor
def floor_detect(floors_total,floor):
    try: 
        if floors_total == floor:
            return 'последний'
        if 1 < floor < floors_total:
            return 'другой'
        if floor == 1:
            return 'первый'
        if floors_total == 0:
            return 'unknown'
    except:
        return 0            
```

```python
# применим функцию к датафрейму в новом столбце floors
df['floors_num'] = df.apply(lambda row: floor_detect(row['floors_total'], row['floor']), axis=1)
```

```python
# выведем количество уникальных значений нового столбца
display(df['floors_num'].value_counts())
```

<br>3. Удаленность от центра. Влияние на цену по данному параметру обнаружено в целом по Спб, нежели по центральному району. Данный параметр может быть применем только к квартирам, территориально находящимся в Санкт-Петербурге.

Спекулятивным является параметр "номер этажа", т.к. большинство квартир находятся в промежутке между 1 и последним этажем в доме, и сложно сказать, как этаж влияет на цену квартиры. В рамках данного исследования никакой зависимости здесь не обнаружено. Также параметр "дата размещения объявления" не оказывает влияния на цену квартиры.
